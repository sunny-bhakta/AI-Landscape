### *==Mathematical operations==*

````python
# Create a tensor of ones by adding 1 to a zeros tensor
ones = torch.zeros(2, 2) + 1
ones
````

````python
# Multiply a tensor of ones by 2 to get a tensor of twos
twos = torch.ones(2, 2) * 2
twos
````

````python
# Arithmetic operations on tensors
# Multiply by 7, subtract 1, then divide by 2
result1 = torch.ones(2, 2) * 7 - 1
result2 = result1 / 2
result1, result2
````

````python
fours = twos ** 2
sqrt2s = twos ** 0.5
fours, sqrt2s
````

````python
# Element-wise addition and subtraction of tensors
a = torch.rand(2, 3)
b = torch.rand(2, 3)
sum_ab = a + b
diff_ab = a - b
a, b, sum_ab, diff_ab
````

````python
s = torch.tensor([1, -2, 3, -4])
s, torch.abs(s), torch.neg(s)
````

````python
# Rounding operations on tensors
m = torch.tensor([1.9, 2.3, 3.7, 4.4])
ceil_m = torch.ceil(m)
round_m = torch.round(m)
floor_m = torch.floor(m)
m, ceil_m, round_m, floor_m
````

````python
# Clamp tensor values to a specified range
# All values below 2 become 2, all above 3 become 3
clamped = torch.clamp(m, min=2, max=3)
clamped
````

### *==Element wise operation==*

````python
# Create a random integer tensor and compute sum, mean, and median
z = torch.randint(size=(2, 3), low=0, high=10, dtype=torch.float32)
sum_z = torch.sum(z)
mean_z = torch.mean(z)
median_z = torch.median(z)
z, sum_z, mean_z, median_z
````

````python
# Mean along different dimensions
mean_all = torch.mean(z)
mean_dim0 = torch.mean(z, dim=0)
mean_dim1 = torch.mean(z, dim=1)
mean_all, mean_dim0, mean_dim1
````

````python
torch.max(z), torch.min(z)
````

````python
# If z = torch.tensor([2, 3, 4]), then torch.prod(z) returns 24 (because 2 × 3 × 4 = 24).
prod_z = torch.prod(z)
prod_z
````

### *Formula For std*
$$\text{std}(x) = \sqrt{\frac{1}{n-1} \sum_{i=1}^n (x_i - \mu)^2}$$

````python
#The standard deviation (std) in PyTorch is calculated as the square root of the average of the squared differences from the mean
torch.std(z)
````

  ### *Formula for Variance*

$$
\text{var}(x) = \frac{1}{n-1} \sum_{i=1}^n (x_i - \mu)^2
$$

Where:
- $x_i$ are the elements of the tensor
- $\mu$ is the mean of the tensor
- $n$ is the number of elements

````python
# - Variance measures how far the values in the tensor are spread out from the mean.
torch.var(z)
````

````python
#Returns the indices of the maximum value of all elements in the :attr:`input` tensor.
torch.argmax(z), torch.argmin(z),
````

### torch.det: Determinant of a Matrix
- For a 2x2 matrix $A = \begin{bmatrix} a & b \\ c & d \end{bmatrix}$, $\text{det}(A) = ad - bc$.
```python
matrix = torch.tensor([[1., 2.], [3., 4.]])
det = torch.det(matrix)  # Returns -2.0
```

````python
torch.det(int_float_mat)
````

#### How torch.inverse Calculates the Inverse
- For a 2x2 matrix $A = \begin{bmatrix} a & b \\ c & d \end{bmatrix}$, the inverse is:
$$
A^{-1} = \frac{1}{ad-bc} \begin{bmatrix} d & -b \\ -c & a \end{bmatrix}
$$

````python
torch.inverse(int_float_mat)
````
